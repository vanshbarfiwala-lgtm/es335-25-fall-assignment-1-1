```python
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
from base import DecisionTree
from metrics import *
from sklearn.model_selection import train_test_split

np.random.seed(42)
num_average_time = 100
```

```python

N = [200,300,400,500]
M = [2, 4, 6, 8]
```

DISCRETE INPUT DISCRETE OUTPUT

```python
train_time_matrix = []
pred_time_matrix = []

for i in N:
    train_row = []
    pred_row = []
    for j in M:
        training_time = []
        pred_time = []
        for k in range(num_average_time):
            x = pd.DataFrame(np.random.randint(0, 3, (i, j)))
            y = pd.Series(np.random.randint(0, 3, i))
            x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
            tree = DecisionTree(criterion='entropy', max_depth=5)

            start_training_time = time.time()
            tree.fit(x_train, y_train)
            end_training_time = time.time()

            start_pred_time = time.time()
            y_pred = tree.predict(x_test)
            end_pred_time = time.time()

            training_time.append(end_training_time - start_training_time)
            pred_time.append(end_pred_time - start_pred_time)

        avg_train = sum(training_time)/len(training_time)
        avg_pred = sum(pred_time)/len(pred_time)

        train_row.append(avg_train)
        pred_row.append(avg_pred)
        print("No. of samples(N)", i)
        print("No. of features(M)", j)
        print("Average training time:",avg_train)
        print("Average prediction time:",avg_pred)

    train_time_matrix.append(train_row)
    pred_time_matrix.append(pred_row)

# Plotting
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(N, [row[0] for row in train_time_matrix], marker='o', label='Training Time')
plt.plot(N, [row[0] for row in pred_time_matrix], marker='s', label='Prediction Time')
plt.title('Time vs N (M={})'.format(M[0]))
plt.xlabel('Number of Samples (N)')
plt.ylabel('Time (seconds)')
plt.legend()

plt.subplot(1,2,2)
plt.plot(M, train_time_matrix[0], marker='o', label='Training Time')
plt.plot(M, pred_time_matrix[0], marker='s', label='Prediction Time')
plt.title('Time vs M (N={})'.format(N[0]))
plt.xlabel('Number of Features (M)')
plt.ylabel('Time (seconds)')
plt.legend()
plt.tight_layout()
plt.show()
```

OBSERVATION:
The training time increases roghly linearly with no. of features and no. of samples in all the cases except for Discrete input and Discrete output.
The prediction time increases roughly with no. of samples.

Theoretically, the time complexity for training time is O(N * M * Log(N)) and for prediction time is O(Log N) since the decision tree is made using recursions. (N here is number of samples and M is number of features)
However in some cases, the graph obtained is different from the expected one. This is likely due to small datasets and the randomness in the data points
